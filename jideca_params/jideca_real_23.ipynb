```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_ari_etc = pd.read_csv('../result/jideca_ari_means_b01_10_23.csv')
df_nmi_etc = pd.read_csv('../result/jideca_nmi_means_b01_10_23.csv')
df_purity_etc = pd.read_csv('../result/jideca_purity_means_b01_10_23.csv')

df_ari_b05 = pd.read_csv('../result/jideca_ari_means_b05_23.csv')
df_nmi_b05 = pd.read_csv('../result/jideca_nmi_means_b05_23.csv')
df_purity_b05 = pd.read_csv('../result/jideca_purity_means_b05_23.csv')
```

```python
df_ari_etc = df_ari_etc.loc[df_ari_etc['types'] == 'real'].sort_values(by='parameters')
df_nmi_etc = df_nmi_etc.loc[df_nmi_etc['types'] == 'real'].sort_values(by='parameters')
df_purity_etc = df_purity_etc.loc[df_purity_etc['types'] == 'real'].sort_values(by='parameters')

df_ari_b05 = df_ari_b05.loc[df_ari_b05['types'] == 'real'].sort_values(by='simple')
df_nmi_b05 = df_nmi_b05.loc[df_nmi_b05['types'] == 'real'].sort_values(by='simple')
df_purity_b05 = df_purity_b05.loc[df_purity_b05['types'] == 'real'].sort_values(by='simple')
```

```python
df_ari_b01 = df_ari_etc.iloc[:5, :].sort_values(by='simple')
df_ari_b02 = df_ari_etc.iloc[5:10, :].sort_values(by='simple')
df_ari_b1 = df_ari_etc.iloc[10:15, :].sort_values(by='simple')
df_ari_b10 = df_ari_etc.iloc[15:20, :].sort_values(by='simple')

df_nmi_b01 = df_nmi_etc.iloc[:5, :].sort_values(by='simple')
df_nmi_b02 = df_nmi_etc.iloc[5:10, :].sort_values(by='simple')
df_nmi_b1 = df_nmi_etc.iloc[10:15, :].sort_values(by='simple')
df_nmi_b10 = df_nmi_etc.iloc[15:20, :].sort_values(by='simple')

df_purity_b01 = df_purity_etc.iloc[:5, :].sort_values(by='simple')
df_purity_b02 = df_purity_etc.iloc[5:10, :].sort_values(by='simple')
df_purity_b1 = df_purity_etc.iloc[10:15, :].sort_values(by='simple')
df_purity_b10 = df_purity_etc.iloc[15:20, :].sort_values(by='simple')
```

```python
df_ari_b01['types'] = r'$\beta$=0.1'
df_ari_b01.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_ari_b02['types'] = r'$\beta$=0.2'
df_ari_b02.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_ari_b05['types'] = r'$\beta$=0.5'
df_ari_b05.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_ari_b1['types'] = r'$\beta$=1'
df_ari_b1.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_ari_b10['types'] = r'$\beta$=10'
df_ari_b10.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)


```

```python
df_nmi_b01['types'] = r'$\beta$=0.1'
df_nmi_b01.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_nmi_b02['types'] = r'$\beta$=0.2'
df_nmi_b02.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_nmi_b05['types'] = r'$\beta$=0.5'
df_nmi_b05.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_nmi_b1['types'] = r'$\beta$=1'
df_nmi_b1.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_nmi_b10['types'] = r'$\beta$=10'
df_nmi_b10.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)


```

```python
df_purity_b01['types'] = r'$\beta$=0.1'
df_purity_b01.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_purity_b02['types'] = r'$\beta$=0.2'
df_purity_b02.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_purity_b05['types'] = r'$\beta$=0.5'
df_purity_b05.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_purity_b1['types'] = r'$\beta$=1'
df_purity_b1.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_purity_b10['types'] = r'$\beta$=10'
df_purity_b10.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
```

```python
df_ari = pd.concat([df_ari_b01, df_ari_b02, df_ari_b05, df_ari_b1, df_ari_b10], ignore_index=True)
df_ari = df_ari.astype({'simple':'str'})

df_nmi = pd.concat([df_nmi_b01, df_nmi_b02, df_nmi_b05, df_nmi_b1, df_nmi_b10], ignore_index=True)
df_nmi = df_nmi.astype({'simple':'str'})

df_purity = pd.concat([df_purity_b01, df_purity_b02, df_purity_b05, df_purity_b1, df_purity_b10], ignore_index=True)
df_purity = df_purity.astype({'simple':'str'})
```

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 14

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
plt.figure(figsize=(8,5))
sns.set_style("ticks", {'axes.grid' : True})
#colors = ['#455E85', '#D68835', '#466B3D']
sns.lineplot(x='simple', y='ari', data=df_ari, linewidth=3, hue='types', style='types', markers=True, markersize=14)
for i in df_ari.groupby('types'):
    for x,y,m in i[1][['simple','ari','ari']].values:
        if x == '0.1': i = 0
        elif x == '0.2': i = 1
        elif x == '0.5': i = 2
        elif x == '1.0': i = 3
        else: i = 4
        #plt.text(i-0.2 , y + 0.01, m)
plt.ylim(0, 0.4)
plt.ylabel('ARI', fontsize=16)
plt.xlabel(r'$\gamma$')
plt.legend(loc='best')
```
