```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_ari_lambda = pd.read_csv('../result/jideca_ari_means_lambda.csv')
df_nmi_lambda = pd.read_csv('../result/jideca_nmi_means_lambda.csv')
df_purity_lambda = pd.read_csv('../result/jideca_purity_means_lambda.csv')
```

```python

df_ari_34 = df_ari_lambda.loc[df_ari_lambda['class'] == 'R34'].sort_values(by='simple')
df_nmi_34 = df_nmi_lambda.loc[df_nmi_lambda['class'] == 'R34'].sort_values(by='simple')
df_purity_34 = df_purity_lambda.loc[df_purity_lambda['class'] == 'R34'].sort_values(by='simple')

df_ari_23 = df_ari_lambda.loc[df_ari_lambda['class'] == 'C23'].sort_values(by='simple')
df_nmi_23 = df_nmi_lambda.loc[df_nmi_lambda['class'] == 'C23'].sort_values(by='simple')
df_purity_23 = df_purity_lambda.loc[df_purity_lambda['class'] == 'C23'].sort_values(by='simple')
```

```python
df_ari_34['types'] = r'$\beta$=0.2, $\gamma$=0.1 (R34)'
df_ari_34.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_ari_23['types'] = r'$\beta$=0.5, $\gamma$=0.1 (C23)'
df_ari_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)

df_nmi_34['types'] = r'$\beta$=0.2, $\gamma$=0.1 (R34)'
df_nmi_34.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_nmi_23['types'] = r'$\beta$=0.5, $\gamma$=0.1 (C23)'
df_nmi_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)

df_purity_34['types'] = r'$\beta$=0.2, $\gamma$=0.1 (R34)'
df_purity_34.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_purity_23['types'] = r'$\beta$=0.5, $\gamma$=0.1 (C23)'
df_purity_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)

```

```python
df_ari = pd.concat([df_ari_34, df_ari_23], ignore_index=True)
df_ari = df_ari.astype({'simple':'str'})

df_nmi = pd.concat([df_nmi_34, df_nmi_23], ignore_index=True)
df_nmi = df_nmi.astype({'simple':'str'})

df_purity = pd.concat([df_purity_34, df_purity_23], ignore_index=True)
df_purity = df_purity.astype({'simple':'str'})

```
