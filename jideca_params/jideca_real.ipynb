```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_ari_b02 = pd.read_csv('../result/jideca_ari_means_b02.csv')
df_nmi_b02 = pd.read_csv('../result/jideca_nmi_means_b02.csv')
df_purity_b02 = pd.read_csv('../result/jideca_purity_means_b02.csv')

df_ari_b05_23 = pd.read_csv('../result/jideca_ari_means_b05_23.csv')
df_nmi_b05_23 = pd.read_csv('../result/jideca_nmi_means_b05_23.csv')
df_purity_b05_23 = pd.read_csv('../result/jideca_purity_means_b05_23.csv')

df_ari_g01 = pd.read_csv('../result/jideca_ari_means_g01.csv')
df_nmi_g01 = pd.read_csv('../result/jideca_nmi_means_g01.csv')
df_purity_g01 = pd.read_csv('../result/jideca_purity_means_g01.csv')

df_ari_g01_23 = pd.read_csv('../result/jideca_ari_means_g01_23.csv')
df_nmi_g01_23 = pd.read_csv('../result/jideca_nmi_means_g01_23.csv')
df_purity_g01_23 = pd.read_csv('../result/jideca_purity_means_g01_23.csv')
```

```python
df_ari_b02 = df_ari_b02.sort_values(by='simple')
df_nmi_b02 = df_nmi_b02.sort_values(by='simple')
df_purity_b02 = df_purity_b02.sort_values(by='simple')

df_ari_b05_23 = df_ari_b05_23.loc[df_ari_b05_23['types'] == 'real'].sort_values(by='simple')
df_nmi_b05_23 = df_nmi_b05_23.loc[df_nmi_b05_23['types'] == 'real'].sort_values(by='simple')
df_purity_b05_23 = df_purity_b05_23.loc[df_purity_b05_23['types'] == 'real'].sort_values(by='simple')

df_ari_g01 = df_ari_g01.loc[df_ari_g01['types'] == 'real'].sort_values(by='simple')
df_nmi_g01 = df_nmi_g01.loc[df_nmi_g01['types'] == 'real'].sort_values(by='simple')
df_purity_g01 = df_purity_g01.loc[df_purity_g01['types'] == 'real'].sort_values(by='simple')

df_ari_g01_23 = df_ari_g01_23.loc[df_ari_g01_23['types'] == 'real'].sort_values(by='simple')
df_nmi_g01_23 = df_nmi_g01_23.loc[df_nmi_g01_23['types'] == 'real'].sort_values(by='simple')
df_purity_g01_23 = df_purity_g01_23.loc[df_purity_g01_23['types'] == 'real'].sort_values(by='simple')
```

```python
df_ari_b02['types'] = r'$\beta$=0.2 (R34)'
df_ari_b02.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_ari_b05_23['types'] = r'$\beta$=0.5 (C23)'
df_ari_b05_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)

df_nmi_b02['types'] = r'$\beta$=0.2 (R34)'
df_nmi_b02.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_nmi_b05_23['types'] = r'$\beta$=0.5 (C23)'
df_nmi_b05_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)

df_purity_b02['types'] = r'$\beta$=0.2 (R34)'
df_purity_b02.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_purity_b05_23['types'] = r'$\beta$=0.5 (C23)'
df_purity_b05_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)

df_ari_g01['types'] = r'$\gamma$=0.1 (R34)'
df_ari_g01.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_ari_g01_23['types'] = r'$\gamma$=0.1 (C23)'
df_ari_g01_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)

df_nmi_g01['types'] = r'$\gamma$=0.1 (R34)'
df_nmi_g01.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_nmi_g01_23['types'] = r'$\gamma$=0.1 (C23)'
df_nmi_g01_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)

df_purity_g01['types'] = r'$\gamma$=0.1 (R34)'
df_purity_g01.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
df_purity_g01_23['types'] = r'$\gamma$=0.1 (C23)'
df_purity_g01_23.drop(['Unnamed: 0', 'parameters'], axis=1, inplace=True)
```

```python
df_ari = pd.concat([df_ari_b02, df_ari_b05_23], ignore_index=True)
df_ari = df_ari.astype({'simple':'str'})

df_nmi = pd.concat([df_nmi_b02, df_nmi_b05_23], ignore_index=True)
df_nmi = df_nmi.astype({'simple':'str'})

df_purity = pd.concat([df_purity_b02, df_purity_b05_23], ignore_index=True)
df_purity = df_purity.astype({'simple':'str'})

df_ari_g = pd.concat([df_ari_g01, df_ari_g01_23], ignore_index=True)
df_ari_g = df_ari_g.astype({'simple':'str'})

df_nmi_g = pd.concat([df_nmi_g01, df_nmi_g01_23], ignore_index=True)
df_nmi_g = df_nmi_g.astype({'simple':'str'})

df_purity_g = pd.concat([df_purity_g01, df_purity_g01_23], ignore_index=True)
df_purity_g = df_purity_g.astype({'simple':'str'})
```

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 14

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
plt.figure(figsize=(8,5))
sns.set_style("ticks", {'axes.grid' : True})
colors = ['#455E85', '#D68835']
sns.lineplot(x='simple', y='ari', data=df_ari, linewidth=3, palette=colors, hue='types', style='types', markers=True, markersize=14)
for i in df_ari.groupby('types'):
    for x,y,m in i[1][['simple','ari','ari']].values:
        if x == '0.1': i = 0
        elif x == '0.2': i = 1
        elif x == '0.5': i = 2
        elif x == '1.0': i = 3
        else: i = 4
        plt.text(i-0.2 , y + 0.01, m)
plt.ylim(0, 0.4)
plt.ylabel('ARI', fontsize=16)
plt.xlabel(r'$\gamma$')
plt.legend(loc='best')
```
